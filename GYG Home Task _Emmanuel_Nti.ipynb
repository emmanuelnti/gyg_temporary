```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline
import warnings
import datetime
warnings.filterwarnings("ignore")


def data_preprocessing (task_1_data_path,task_2_data_path):
    data_task1 = pd.read_csv(task_1_data_path)
    data_task2 = pd.read_csv(task_2_data_path)
    
    print("Initial data size:", data_task1.shape[0])
    print("Initial data size:",data_task2.shape[0])
    data_task1['date_of_checkout'] =  pd.to_datetime(data_task1['date_of_checkout'], infer_datetime_format=True, utc=True )
    data_task1['date_of_travel'] =  pd.to_datetime(data_task1['date_of_travel'],  infer_datetime_format=True, utc=True )
    data_task2['date_of_creation'] = pd.to_datetime(data_task2['date_of_creation'],  infer_datetime_format=True,utc=True )
    data_task2['date_of_checkout'] = pd.to_datetime(data_task2['date_of_checkout'],  infer_datetime_format=True ,utc=True)
    data_task2['update_timestamp'] = pd.to_datetime(data_task2['update_timestamp'],  infer_datetime_format=True ,utc=True)
    data_task2['cycle_timestamp'] = pd.to_datetime(data_task2['cycle_timestamp'],  infer_datetime_format=True ,utc=True)
    data_task2['week'] = pd.to_datetime(data_task2['week'],  infer_datetime_format=True, utc=True) 
    data_task2.drop_duplicates(inplace=True)
    data_task1.drop_duplicates(inplace=True)
    data_task1.reset_index(inplace=True, drop=True)
    data_task2.reset_index(inplace=True, drop=True)
    return  data_task1, data_task2
    
data_task1,data_task2 = data_preprocessing("data_task1.csv","data_task2.csv")


# Displaying the data
print()
print ('\033[1m  Task 1 Data \033[0m')
print()

data_task1.info()
display(data_task1.head(5))
display(data_task1.describe().T)

print()
print ('\033[1m  Task 2 Data \033[0m')
print()
data_task2.info()
display(data_task2.sample(5))
display(data_task2.describe().T)
```

## Task 4

- Using the fact_checkout table below, the SQL query below would generate  a list of customer_ids and first checkout 
date for customers who have made their second checkout within 30 days of their first check out:

       
WITH
    metrics AS (SELECT DATE_TRUNC('day', date_of_checkout) AS checkout_date 
    ,customer_id_a AS customer_id
FROM fact_checkout),

adding_window AS (SELECT customer_id 
    ,checkout_date
    ,min(checkout_date) over (partition by customer_id order by checkout_date) as first_checkout_date
FROM metrics),

get_daily_cohort AS (SELECT customer_id
    ,checkout_date
    ,first_checkout_date
    ,DATEDIFF(checkout_date
    ,first_checkout_date) AS cohort_day
FROM adding_window)

SELECT first_checkout_date
    ,customer_id
FROM get_daily_cohort 
WHERE cohort_day >0 AND cohort_day <30;
